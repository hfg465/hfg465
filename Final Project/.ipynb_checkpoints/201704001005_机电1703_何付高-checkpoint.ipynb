```python
##1）将白酒质量数据集进行数据探索，分析白酒质量与各个特征之间的关系，并进行可视化。
##2）建立机器学习模型进行白酒质量预测，要求划分训练集和验证集，并进行5折交叉验证。如果验证集预测平均精确率在80%以上，则给与优秀成绩。
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.ticker
from matplotlib import rcParams
import warnings
warnings.filterwarnings('ignore')
```

```python
##定义 Seaborn 全局绘图参数
sns.set()
sns.set_context(
    "notebook",
    font_scale=1.5,
    rc={
        "figure.figsize": (11, 8),
        "axes.titlesize": 18
    }
)

rcParams['figure.figsize'] = 11, 8
```

```python
df = pd.read_csv(
    'winequality-white.csv', sep=';')
print('Dataset size: ', df.shape)
df.head()##预览数据集
```

```python
查看不同质量评分白酒的数量
```

```python
qn=df['quality'].value_counts()
print(qn)
```

```python
features = ['quality']
df[features].hist(figsize=(8, 4))
```

```python
由直方图可知，白酒质量呈正态分布，下面查找每个特征之间的相关性，尤其quality与其他特征的相关性
```

```python
numerical = list(set(df.columns))
corr_matrix = df[numerical].corr()
sns.heatmap(corr_matrix)
```

```python
    由可视化结果可以看出，quality与free sulfur dioxide,sulphates,pH,citric acid几个特征几乎没有线性关系，与其他特征之间的线性关系也不明显
可见白酒质量与相关因素没有单纯的线性关系，而是由这些因素共同影响
```
